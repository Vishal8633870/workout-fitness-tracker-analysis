```python
import pandas as pd
import os
import kagglehub

dataset_path = kagglehub.dataset_download("adilshamim8/workout-and-fitness-tracker-data")

df = pd.read_csv(os.path.join(dataset_path, "workout_fitness_tracker_data.csv"))

df.head()

df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df["Workout Duration (mins)"], bins=30, kde=True)
plt.title("Distribution of Workout Duration")
plt.xlabel("Workout Duration (mins)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["Workout Duration (mins)"], y=df["Calories Burned"])
plt.title("Calories Burned vs. Workout Duration")
plt.xlabel("Workout Duration (mins)")
plt.ylabel("Calories Burned")
plt.show()
```

```python
df.groupby("Workout Type")["Calories Burned"].mean().sort_values().plot(kind="barh", figsize=(10,6), title="Average Calories Burned by Workout Type")
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["Workout Intensity"], y=df["Heart Rate (bpm)"])
plt.title("Heart Rate Distribution by Workout Intensity")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["Sleep Hours"], y=df["Workout Duration (mins)"])
plt.title("Sleep Hours vs. Workout Duration")
plt.xlabel("Sleep Hours")
plt.ylabel("Workout Duration (mins)")
plt.show()
```
